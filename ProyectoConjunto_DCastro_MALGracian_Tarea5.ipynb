Para demostrar la validez de la ecuación dada, podemos utilizar inducción matemática.

**Paso base:** Para i = 0, la ecuación se convierte en:
$$p_1 = \frac{n}{1}\frac{p}{1-p} p_0$$
Lo cual es equivalente a $p_1 = \frac{n}{1}\frac{p}{1-p} p_0$, que es verdadero.

**Paso de inducción:** Supongamos que la ecuación es válida para algún valor k, es decir:
$$p_{k+1} = \frac{n-k}{k+1}\frac{p}{1-p} p_k$$

**Paso inductivo:** Consideremos i = k + 1:
$$p_{k+2} = \frac{n-(k+1)}{(k+1)+1}\frac{p}{1-p} p_{k+1}$$
Esto se simplifica a:
$$p_{k+2} = \frac{n-k-1}{k+2}\frac{p}{1-p} \left(\frac{n-k}{k+1}\frac{p}{1-p} p_k\right)$$
$$p_{k+2} = \frac{n-k-1}{k+2}\frac{n-k}{k+1}\frac{p}{1-p} \frac{p}{1-p} p_k$$
$$p_{k+2} = \frac{(n-k)(n-k-1)}{(k+1)(k+2)}\frac{p^2}{(1-p)^2} p_k$$
Esto es exactamente la forma de la ecuación para i = k + 1, por lo tanto, se mantiene válida para este caso.

Por lo tanto, la ecuación es válida para todos los enteros no negativos i por inducción matemática.

\begin{align*}
\frac{P(k+1)}{P(k)} &= \frac{\frac{e^{-\lambda}\lambda^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}\lambda^k}{k!}} \\
&= \frac{e^{-\lambda}\lambda^{k+1}}{(k+1)!} \cdot \frac{k!}{e^{-\lambda}\lambda^k} \\
&= \frac{e^{-\lambda}\lambda^{k+1}}{e^{-\lambda}\lambda^k} \cdot \frac{k!}{(k+1)!} \\

### 2. Función de densidad $(g(x))$

Dado el intervalo $( (0,1) )$ y la función de densidad $( f(x) = 30x^2(1-x)^3 )$ para $( x \in (0,1) )$, podemos encontrar la función de densidad $( g(x) )$

Primero, recordemos que la función de densidad $( g(x) )$ se obtiene dividiendo $( f(x) )$ por su integral en el dominio $( (0,1) )$:

$[ g(x) = \frac{f(x)}{\int_{0}^{1} f(u) du} ]$

Calculemos la integral $( \int_{0}^{1} f(u) du )$:

$$ \int_{0}^{1} 30u^2(1-u)^3 du \$$

Usando la regla del producto y la regla de potencias, podemos simplificar la integral. Realicemos los cálculos:

$$\int_{0}^{1} 30u^2(1-u)^3 du = \int_{0}^{1} 30u^2(1-3u+3u^2-u^3) du$$

$$= \int_{0}^{1} 30u^2 - 90u^3 + 90u^4 - 30u^5 du$$

$$= \left( 10u^3 - \frac{45}{2}u^4 + 18u^5 - \frac{15}{2}u^6 \right)_{0}^{1}$$

$$= 10 - \frac{45}{2} + 18 - \frac{15}{2} = 10$$

Ahora, podemos escribir la función de densidad $( g(x) )$:

$$[ g(x) = \frac{30x^2(1-x)^3}{10} = 3x^2(1-x)^3 ]$$

Por lo tanto, la función de densidad $( g(x) )$ es $( 3x^2(1-x)^3 )$.

### 8. Demostración de la Ecuación de Recurrencia Binomial

Dada la ecuación de recurrencia para la distribución binomial:

$$
p_{i+1} = \frac{n - i}{i + 1} \cdot \frac{p}{1 - p} \cdot p_i
$$

Queremos demostrar su validez. Comencemos con la derivación paso a paso.

1. **Base de Inducción**: Para $(i = 0)$, la ecuación es trivialmente verdadera porque $(p_1 = p_0 \cdot \frac{n}{1} \cdot \frac{p}{1 - p})$.

2. **Supongamos que es Verdadero para $i$**: Supongamos que la ecuación es válida para algún $i = k$, es decir, $p_{k+1} = \frac{n - k}{k + 1} \cdot \frac{p}{1 - p} \cdot p_k$.

3. **Demostrar para $i = k + 1$**: Ahora, veamos si es válido para $i = k + 1$.

$$
\begin{aligned}
p_{k+2} & = \frac{n - (k + 1)}{k + 2} \cdot \frac{p}{1 - p} \cdot p_{k+1} \\
& = \frac{n - (k + 1)}{k + 2} \cdot \frac{p}{1 - p} \cdot \left(\frac{n - k}{k + 1} \cdot \frac{p}{1 - p} \cdot p_k\right) \quad \text{(por suposición inductiva)} \\
& = \frac{(n - k)(n - (k + 1))}{(k + 1)(k + 2)} \cdot \frac{p^2}{(1 - p)^2} \cdot p_k \\
& = \frac{n - (k + 1)}{k + 2} \cdot \frac{p}{1 - p} \cdot p_{k+1}
\end{aligned}
$$

Por lo tanto, hemos demostrado la ecuación de recurrencia binomial. La base de inducción y el paso inductivo confirman su validez para todos los $i$.

### 9. Generación de Variables Aleatorias Poisson y Demostración Matemática 

$$\frac{p_{k+1}}{p_k} = \frac{e^{-\lambda}\lambda^k}{k!}$$

Multiplicamos ambos lados por $( k+1 )$:

$$\frac{p_{k+1}(k+1)}{p_k} = \frac{e^{-\lambda}\lambda^k}{(k-1)!}$$

Simplificamos:

$$\frac{p_{k+1}(k+1)}{p_k} = \frac{\lambda}{k} \frac{e^{-\lambda}\lambda^{k-1}}{(k-1)!}$$

Esta expresión coincide con la relación que queríamos demostrar:

$$p_{k+1} = \frac{\lambda}{k+1} p_k$$

Por lo tanto, hemos demostrado matemáticamente que $P(k+1) = \frac{\lambda}{k+1} P(k)$ en el contexto de la distribución de Poisson.
